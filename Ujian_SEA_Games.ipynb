```python
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
import requests
import matplotlib.pyplot as plt
```

```python
web=requests.get('https://www2.2019seagames.com/countries/')
data=BeautifulSoup(web.content,'html.parser')
```

```python
seag17=[]
for i in data.findAll('em'):
    seag17.append(i.string)

negara=[seag17[0],seag17[6],seag17[12],seag17[18],seag17[24],seag17[30],seag17[36],seag17[42],seag17[48],seag17[54],seag17[60]]

y=[int(seag17[2].replace('Gold:','')),int(seag17[8].replace('Gold:','')),int(seag17[14].replace('Gold:','')),int(seag17[20].replace('Gold:','')),int(seag17[26].replace('Gold:','')),int(seag17[32].replace('Gold:','')),int(seag17[38].replace('Gold:','')),int(seag17[44].replace('Gold:','')),int(seag17[50].replace('Gold:','')),int(seag17[56].replace('Gold:','')),int(seag17[62].replace('Gold:',''))]

z=[2,4,72,1,55,4,149,53,92,0,98]
```

```python
plt.style.use('seaborn')
plt.figure(figsize=(16,8))
plt.plot(negara,y,'ro-',label='Gold 2017')
plt.plot(negara,z,'go-',label='Gold 2019')
plt.scatter([4,6],[145,149],marker='*',color='goldenrod',s=500,label='Champion')
plt.title('GOLD MEDAL TALLY SEAG GAMES 2017 & 2019')
plt.legend()
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(20,8))
plt.subplot(121)
plt.pie(y,labels=negara,autopct='%1.1f%%')
plt.title('Gold % SEA Games 2019')

plt.subplot(122)
plt.pie(z,labels=negara,autopct='%1.1f%%')
plt.title('Gold % SEA Games 2019')

plt.show()
```
